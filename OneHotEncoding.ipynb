One Hot encoding: Converting categorical data into numerical data

```python
import pandas as pd
```

```python
dt=pd.read_csv("loan.csv")
dt.head(3)
```

```python
dt.isnull().sum()
```

```python
dt["Gender"].fillna(dt["Gender"].mode()[0],inplace=True)
dt["Married"].fillna(dt["Married"].mode()[0],inplace=True)
```

```python
en_data=dt[["Gender","Married"]]
```

```python
pd.get_dummies(en_data).info()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop="first")
ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ohe.fit_transform(en_data).toarray(),columns=["Gender_Male","Married_Yes"])
```
